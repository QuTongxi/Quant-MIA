```python
import matplotlib.pyplot as plt

a_x = [62.35, 68.10, 70.05]
a_y = [0.0028, 0.2009, 0.3712]

b_x = [68.10, 69.64, 70.04]
b_y = [0.0011, 0.2934, 0.3650]

obc_x = [65.47, 68.07, 70.05]
obc_y = [0.0189, 0.2288, 0.3647]

plt.figure(figsize=(4.0, 3))

colors = ["tab:green", "tab:orange", "tab:purple"]
markers = [".", "o", "D"]
s = 70

plt.plot(a_x, a_y, color=colors[0], linestyle="--", label="AdaRound")
plt.plot(b_x, b_y, color=colors[1], linestyle="--", label="BRECQ")
plt.plot(obc_x, obc_y, color=colors[2], linestyle="--", label="OBC")

for i in range(3):
    plt.scatter(
        a_x[i],
        a_y[i],
        color=colors[0],
        marker=markers[i],
        edgecolors="black",
        linewidths=1,
        s=s,
        label="1.58-bit" if i == 0 else "",
    )
    plt.scatter(
        b_x[i],
        b_y[i],
        color=colors[1],
        marker=markers[i],
        edgecolors="black",
        linewidths=1,
        s=s,
        label="2-bit" if i == 1 else "",
    )
    plt.scatter(
        obc_x[i],
        obc_y[i],
        color=colors[2],
        marker=markers[i],
        edgecolors="black",
        linewidths=1,
        s=s,
        label="4-bit" if i == 2 else "",
    )

plt.scatter(
    [70.05],
    [0.3683],
    color="tab:red",
    marker="*",
    s=500,
    zorder=0,
    edgecolors="black",
    linewidths=1,
    label="32-bit (full)",
)

plt.xlabel("Test Accuracy (%)")
plt.ylabel("TPR @ 0.1% FPR")
plt.xlim(61, 71)
plt.ylim(7e-4, 1.0)
plt.yscale("log")
plt.grid(axis="both", linestyle="dotted", color="tab:grey", alpha=0.75)

plt.tight_layout()
legend = plt.legend(
    bbox_to_anchor=(0.2, 0.66), loc="center", borderaxespad=0.0, handletextpad=0.3
)
for legend_item in legend.legendHandles[3:]:
    legend_item.set_color("gray")
    legend_item.set_sizes([50])
legend.legendHandles[-1].set_sizes([150])

plt.savefig("teaser.pdf", format="pdf", dpi=300, bbox_inches="tight")
```
